```python
def calculator(num1, num2):
    add = num1 + num2
    sub = num1 - num2
    multiply = num1 * num2
    divide = num1 / num2 if num2 != 0 else "Cannot divide by zero"
    mod = num1 % num2 if num2 != 0 else "Cannot mod by zero"
    print(f"Addition = {add}, Subtraction = {sub}, Multiply = {multiply},Divide = {divide}, Remainder = {mod}");
    return add, sub, multiply, divide, mod
```

```python
calculator(2,3)
```

```python
def housing_cleaning(room_list):
    for room in room_list:
        print("Checking:", room)
        if room == "Kitchen":
            print("Wiping the counters, mopping the floor, and taking out the trash in the kitchen.")
        elif room == "Washroom":
            print("Scrubbing the sink and toilet, cleaning the mirror, and sanitizing the washroom.")
        elif room == "Bedroom":
            print("Making the bed, arranging pillows, and putting things in order in the bedroom.")
        else:
            print(room, "needs general cleaning: dusting surfaces and tidying up the area.")
        print("Cleaning completed\n")
```

```python
housing_cleaning(["Bedroom","Washroom","ShapaterRoom"])
```

```python
# PYTHON FILE MODES - COMPLETE GUIDE

# ==================== FILE MODES ====================

# 'r' - READ MODE (Default)
# Opens file for reading only. File must exist. Cannot write.
# f = open("data.txt", "r")

# 'w' - WRITE MODE
# Opens file for writing. Creates new file or OVERWRITES existing file.
f = open("data.txt", "w")
f.write("Hello World")

# 'a' - APPEND MODE
# Opens file for appending. Adds new data at end. Does not delete old data.
f = open("data.txt", "a")
f.write("\nNew line appended")

# 'x' - CREATE MODE
# Creates NEW file only if it doesn't exist.
# f = open("newfile.txt", "x")

# 'r+' - READ AND WRITE MODE
# Opens file for both reading and writing. File must exist.
# f = open("data.txt", "r+")

# 'w+' - WRITE AND READ MODE
# Opens file for both writing and reading. OVERWRITES existing file.
# f = open("data.txt", "w+")

# 'a+' - APPEND AND READ MODE
# Opens file for both appending and reading. Does not delete old data.
# f = open("data.txt", "a+")

# ==================== FILE METHODS (READING) ====================

# read() - Reads ENTIRE file as ONE string
with open("data.txt", "r") as f:
    entire_content = f.read()  # Returns: "Hello World\nNew line appended"
    print("read():", entire_content)

# readline() - Reads ONE LINE at a time
# Each call reads next line. Includes \n at end.
with open("data.txt", "r") as f:
    line1 = f.readline()  # Returns: "Hello World\n"
    line2 = f.readline()  # Returns: "New line appended"
    print("readline():", line1, line2)

# readlines() - Reads ALL lines as a LIST
# Each line includes \n at end
with open("data.txt", "r") as f:
    all_lines = f.readlines()  # Returns: ["Hello World\n", "New line appended"]
    print("readlines():", all_lines)

# ==================== FILE METHODS (WRITING) ====================

# write() - Writes a SINGLE STRING to file
with open("data.txt", "w") as f:
    f.write("First line")
    f.write("\nSecond line")
    # Result in file: First line\nSecond line

# writelines() - Writes a LIST of strings to file
# Does NOT add \n automatically between items
with open("data.txt", "w") as f:
    lines = ["Hello\n", "World\n", "Python\n"]
    f.writelines(lines)
    # Result in file: Hello\nWorld\nPython\n

# ==================== FILE METHODS (POSITION) ====================

# seek(position) - Moves file pointer to specific position
# 0 = beginning, 1 = current position, 2 = end
with open("data.txt", "r+") as f:
    f.seek(0)  # Move to beginning of file
    content = f.read()
    print("Content from beginning:", content)

# tell() - Returns current position of file pointer
with open("data.txt", "r") as f:
    print("Position:", f.tell())  # Returns: 0
    f.read()
    print("Position after read:", f.tell())  # Returns: file_size

# ==================== FILE METHODS (CLOSING) ====================

# close() - Closes the file (not recommended, use 'with' instead)
f = open("data.txt", "r")
content = f.read()
f.close()  # MUST do this manually

# flush() - Writes buffered data to disk immediately
# Useful when you need to ensure data is saved
f = open("data.txt", "w")
f.write("Important data")
f.flush()  # Forces write to disk before close()
f.close()

# ==================== BEST PRACTICE ====================

# Always use 'with' statement - File closes automatically
with open("data.txt", "r") as f:
    content = f.read()
    print(content)

print("\n=== FILE OPERATIONS SUMMARY ===")
print("read() = Read entire file as string")
print("readline() = Read one line at a time")
print("readlines() = Read all lines as list")
print("write() = Write single string")
print("writelines() = Write list of strings")
print("seek(pos) = Move file pointer to position")
print("tell() = Get current file pointer position")
print("close() = Close file (use 'with' instead)")
print("flush() = Force write to disk")
```
