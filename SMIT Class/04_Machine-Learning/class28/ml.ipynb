### Linear Regression
    Used when we want to predict numeric value

    Simple Linear Regression (One input features and one output features -> features means column)
    Multiple Linear Regression (Many input features and Many output features -> features means column)

### Categorical Classification
    Used to predict categorical data

```python
data.head()
```

```python
x = data[['Height']]
y = data[['Weight']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=44)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt
data.plot(kind='scatter', x='Height', y='Weight', 
       title='Weight and Height in adults')
plt.plot(x_train,y_pred, color='red', linewidth=3)
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
r2_score(y_train,y_pred)
```

```python
mean_absolute_error(y_train,y_pred)
```

```python
y_pred_test = lr.predict(x_test)
```

```python
r2_score(y_test,y_pred_test)
```

```python
mean_absolute_error(y_test,y_pred_test)
```
