# Calculator App

```python
while True:
    input_number_1 = int(input("Enter a First number"))
    if input_number_1 ==00:
        break
    input_number_2 = int(input("Enter a secound number"))
    if input_number_2 == 00:
        break
    ope = input("Enter a operation")
    
    if ope =="+":
        ope_result = input_number_1+input_number_2
    elif ope =="-":
        ope_result = input_number_1-input_number_2
    elif ope =="*":
         ope_result = input_number_1*input_number_2
    elif ope =="/":
        ope_result = input_number_1/input_number_2
    elif ope =="//":
         ope_result = input_number_1//input_number_2
    elif ope =="%":
        ope_result = input_number_1%input_number_2
    elif ope =="**":
         ope_result = input_number_1**input_number_2
    else:
        print("Invalid Operator")
    print(f"""             Calculator App
                         =================
                     input_number_1 : {input_number_1},
                     input_secound_2 : {input_number_2},
                     operator   :       {ope_result}
             _____________________________________________________                   

""")
    
    
```
