**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
# Write your code here for Q2
class BankAccount:
    def __init__(self,balance):
        self.__balance = balance
    def deposit(self,amount):
        if amount <= 0:
            print("Invalid Amount Entered")
        else:
            self.__balance +=amount
            print(f"Amount Sucessfully deposited. Your new balance is {self.__balance + amount}")
    def withdraw(self,amount):
        if amount > self.__balance:
            print("Insufficent amount is entered")
        else:
            self.__balance -=amount
            print(f"Amount withdrawed. Your new balance is {self.__balance - amount}")
obj = BankAccount(19999)
obj.deposit(100)
obj.withdraw(100)
obj.withdraw(1100)

# obj.__balance
        
```

**Q4:** Protected variable (`_`) aur Private variable (`__`) ke beech kya farq hota hai? Aik simple code ke zariye dono ka use-case dikhayein.

```python
# Write your code here for Q4
class Parent:
    def __init__(self):
        self._protected_val = "Protected Data"  # Single underscore
        self.__private_val = "Private Data"      # Double underscore
class Child(Parent):
    def show_data(self):
        # Child class protected variable ko direct access kar sakti hai
        print(f"Child accessing protected: {self._protected_val}")
       
        # Child class private variable ko direct access NAHI kar sakti (Error aayega)
        try:
            print(self.__private_val)
        except AttributeError:
            print("Child CANNOT access private variable directly!")
obj = Child()
obj.show_data()

print("\n--- Testing From Outside ---")
print(f"Outside accessing protected: {obj._protected_val}")
try:
    print(obj.__private_val)
except AttributeError:
    print("Outside CANNOT access private variable directly!")
print(f"Accessing private via Name Mangling: {obj._Parent__private_val}")
```

**Q6:** Single Inheritance ka use karte hue aik parent class `Vehicle` (with brand attribute) aur child class `Car` (with model attribute) banayein. Car ka object bana kar dono attributes print karein.

```python
# Write your code here for Q6
class Vehicle:
    def __init__(self,brand):
        self.brand = brand
    def show(self):
        print(f"Your vehicle brand is {self.brand}")
class Car(Vehicle):
    def __init__(self, brand,model):
        super().__init__(brand)
        self.model = model
    def show(self):
        super().show()
        print(f"Your vehicle model is {self.model}")
obj = Car("Toyota","Corolla")
obj.show()
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
# Write your code here for Q10
class Shape:
    def __init__(self,color):
        self.color = color
    def intro(self):
        print(f"The color of the shape is {self.color}")
class Circle(Shape):
    def __init__(self, color,radius):
        super().__init__(color)
        self.radius = radius
    def area(self):
        print(f"The area of circle is {3.14 * self.radius * self.radius}")
class Square(Shape):
    def __init__(self, color,sides):
        super().__init__(color)
        self.sides = sides
    def area(self):
        print(f"The area of square is {self.sides * self.sides}")
obj = Circle("Red",23)
obj.intro()
obj.area()
obj2 = Square("Blue",14)
obj2.intro()
obj2.area()
```

**Q15:** Kya Python default roop se Method Overloading (same method name with different arguments) support karta hai? Agar nahi, toh hum same name ke method ko multiple arguments ke sath kaise handle karte hain? Code likhein.

```python
# Write your code here for Q15
# by default python doesnt support overloading but we can do it using *args which converts it into a tuple and deal with arguments
class Calculator:
    def add(self,*args):
        return sum(args)
obj = Calculator()
print(obj.add(5))
print(obj.add(5,100))
print(obj.add(5,89,90))
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

```python
# Write your code here for Q17
from abc import ABC,abstractmethod
class Computer(ABC):
    @abstractmethod
    def process(self):
        pass
class Laptop(Computer):
    def __init__(self):
        print("I Am a laptop")
    def process(self):
        print("I am a portable thing")
class Desktop(Computer):
    def __init__(self):
        print("I am a Desktop")
    def process(self):
        print("I Am not portable")
obj = Laptop()
obj.process()
obj2 = Desktop()
obj2.process()
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
# Write your code here for Q20
from abc import ABC,abstractmethod

class Payment(ABC):
    def __init__(self,amount):
        self.amount = amount
    @abstractmethod
    def processPayment(self):
        '''Every payment method must define its own'''
        pass
    def paymentRecipt(self):
        print(f"Payment recieved successfully {self.amount}")
class CreditCardPayment(Payment):
    def __init__(self, amount,cardNumber,cardHolder):
        super().__init__(amount)
        self.cardNumber = cardNumber
        self.cardHolder = cardHolder
    def processPayment(self):
        print(f"Processing your transaction for card number {self.cardNumber}")
        print(f"Card Holder name is {self.cardHolder}")
    def paymentRecipt(self):
        super().paymentRecipt()
        print(f"Payment is processed by card. The holder name is {self.cardHolder}")
class UpiPayment(Payment):
    def __init__(self, amount,upi_id):
        super().__init__(amount)
        self.upi_id = upi_id
    def processPayment(self):
        print(f"Initiating UPI payment request via ID: {self.upi_id}...")
        print("Waiting for secure MPIN authentication...")
        print(f"Rs. {self.amount} successfully transferred via UPI gateway.")
    def paymentRecipt(self):
        super().paymentRecipt()
        print(f"This payment is processed by Upi id-> {self.upi_id}")
        
print("--- Transaction 1: Shopping Mall ---")
card_tx = CreditCardPayment(4500, "1234567890123456", "Shahzaib Baig")
card_tx.processPayment()
card_tx.paymentRecipt()

print("--- Transaction 2: Online QR Scanner ---")
upi_tx = UpiPayment(120, "shahzaib@upi")
upi_tx.processPayment()
upi_tx.paymentRecipt()
```
