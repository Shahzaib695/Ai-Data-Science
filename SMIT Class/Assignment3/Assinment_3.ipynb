#### Q10: Count Vowels in String

```python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"
count = 0

for ch in text:
    if ch in vowels:
        count += 1

print("Number of vowels:", count)
```

#### Q11: Reverse List using Loop

```python
lst = [1, 2, 3, 5, 6]
rlst = []

for i in range(len(lst) - 1, -1, -1):
    rlst.append(lst[i])

print(rlst)
```
