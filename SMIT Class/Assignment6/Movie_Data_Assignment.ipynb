### **Question 15**
'Runtime' column se 'min' text ko hata kar use proper integer (numeric) data type mein convert karein.

```python
df['Runtime'] = df['Runtime'].astype(str).str.replace('min', '', regex=False).str.strip()
df['Runtime'] = df['Runtime'].astype(int)
df['Runtime']
```

### **Question 16**

### **Question 21**

### **Question 22**

### **Question 26**

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
top5_rated = df.sort_values(by='IMDB_Rating', ascending=False).head(5)
x = range(len(top5_rated))
width = 0.35
plt.figure(figsize=(10,6))
plt.bar([i - width/2 for i in x], top5_rated['Meta_score'], width=width, label='Meta_score')
plt.bar([i + width/2 for i in x], top5_rated['Rating_100_Scale'], width=width, label='IMDB_Rating (x10)')
plt.xticks(list(x), top5_rated['Series_Title'], rotation=45, ha='right')
plt.legend()
plt.title('Top 5 Movies: Meta_score vs IMDB_Rating (scaled)')
plt.tight_layout()
plt.show()
```
