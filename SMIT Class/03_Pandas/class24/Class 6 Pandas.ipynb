### Merging

Explanation of `pd.merge(df1, df2, on="sales_man_id", how="outer")`

- `pd.merge(...)` combines two DataFrames using one or more common key columns.
- `df1` and `df2` are the input tables.
- `on="sales_man_id"` means the merge matches rows where `sales_man_id` is equal in both DataFrames.
- `how="outer"` means the result keeps all keys from both `df1` and `df2`.

Purpose of `merge` / join:
- To combine related data from different sources into one table.
- To enrich a dataset by bringing in additional columns from another table that share a common key.
- To perform relational-style joins, similar to SQL joins.

How the `how` options work:

- `how="inner"`
    - Keeps only rows with matching keys in both DataFrames.
    - Result contains the intersection of keys.
    - Rows with keys present in only one table are dropped.

- `how="outer"`
    - Keeps all rows from both DataFrames.
    - Result contains the union of keys.
    - Missing values are filled with `NaN` where a match does not exist.

- `how="left"`
    - Keeps all rows from the left DataFrame (`df1` in this case).
    - Adds matching rows from the right DataFrame where keys match.
    - If no match exists in the right DataFrame, the right-side columns are `NaN`.

- `how="right"`
    - Keeps all rows from the right DataFrame (`df2`).
    - Adds matching rows from the left DataFrame where keys match.
    - If no match exists in the left DataFrame, the left-side columns are `NaN`.

Additional notes:
- If the key column names differ, use `left_on` and `right_on` instead of `on`.
- If there are duplicate keys, merge can produce multiple rows for each matching pair (Cartesian product for that key).
- `pd.concat([...])` is different: it stacks DataFrames by rows or columns, and does not match on key values.

```python
pd.merge(df1,df2, on= "sales_man_id",how= "outer")
```
