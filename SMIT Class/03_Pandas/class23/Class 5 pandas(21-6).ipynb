### Descritization and binning
    numbers
    1) Discrete (Countable) [1,2,3,4,5,6]
    2) Continous (Measurable) [1,3,89.9,9.00]
    continous ko discrete mai lai ate hai

```python
df.Age.min()
```

```python
df.Age.max()
```

```python
bins = [20,30,40,50]
```

```python
pd.cut(df.Age,bins=bins,labels=["Young","Adult","Uncle"])
```

```python
df["new_age_group"] = pd.cut(df.Age,bins=bins,labels=["Young","Adult","Uncle"])
```

```python
df['new_age_group'].value_counts().plot(kind='bar', color='darkblue')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Age Groups')
plt.xticks(rotation=45)
plt.show()
```
