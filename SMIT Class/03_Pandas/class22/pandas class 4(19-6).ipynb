### Mean,Median,Mode

    Categorical data: 
        Designation is a type of categorical data
        We use mode for categorical data. Wherever we spot this type of data we fill it using mode.
    Numerical data:
    - Data in numbers.
    - For numerical data we don’t use mode by default. We use measures of central tendency instead:
        1. Mean = Average
             - Add all numbers ÷ count of numbers.
             - Use it when data is balanced with no big outliers.
             - Ex: Salary data 30k, 35k, 32k → Mean ≈ 32.3k
        2. Median = Middle value
             - Sort data, pick the middle one. If even count, average the 2 middle values.
             - Use it when data has outliers or is skewed.
             - Ex: Salary data 30k, 32k, 1M → Median = 32k, but Mean gets pulled up by 1M
        3. Mode = Most frequent value
             - Works for numerical data too, but only useful if some number repeats a lot.
             - Ex: Test scores 70, 70, 75, 80 → Mode = 70

    In data cleaning/ML, most people use median for numerical missing values because it’s robust to outliers. Mean is used when data is normally distributed.

### Outliers
    [10,20,25,30,40,100] ---> outliers in data
    An outlier is a number in a dataset that is completely different from all the other numbers. It’s the "odd one out" that stands way too far away from the rest of the group.

    In your data: [10, 20, 25, 30, 40, 100]

    Most of the numbers are close together, sitting comfortably between 10 and 40.

    100 is the outlier because it is massively larger than everything else.

    If outlier is far then we will fill data using median
    If outlier is near then we will fill data using mean
