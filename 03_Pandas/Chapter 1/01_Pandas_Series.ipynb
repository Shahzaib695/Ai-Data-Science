### Pandas Series
    A series is a one dimensional vertical array which is used for storing data of any data type. The axis labelled is known as index.

### Creating Series
    There are multiple ways to create series in Pandas

```python
import numpy as np 
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
my_array = np.array([10,20,30])
d = {
    1:10,
    2:20,
    3:30
}
```

```python
pd.Series(my_list)
# the left side is named as labels while the right side elements are known as data point
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(my_array)
# does not support 2d array/Matrix
```

```python
pd.Series(d)
# It uses key as index
```

```python
e = {
    'a':10,
    'b':20,
    'c':30
}
pd.Series(e)
```
