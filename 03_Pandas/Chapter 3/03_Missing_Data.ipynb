```python
import pandas as pd
import numpy as np
```

### Finding Missing Data

```python
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df.isna() #-> false shows not having any value while true shows it has
```

```python
df.isna().sum()
```

```python
df.isna().any()
#this checks does it have any null values in any columns
```

### Removing Missing Data

```python
df
```

```python
df.dropna() #removes null values row
```

```python
# we can also set threshold for removing the values like if we giive 3 as threshold it will check whether the column has alteast 3 non null values if it has it will not remove it otherwise it will remove it
df.dropna(thresh=3)
```

### Filling The Missing Data

```python
df
```

```python
df.fillna(0)
```

```python
values = {
    'A' : 0,
    'B' : 100,
    'C' : 200,
    'D' : 300
}
```

```python
df.fillna(value=values) #column wise selection hota hai
# drop row wise hota hai
```
