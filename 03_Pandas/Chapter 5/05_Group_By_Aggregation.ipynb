```python
import numpy as np
import pandas as pd
```

### Group By

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Group By Category and Calculate the sum of sales
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group By Store And Calculate The Sum Of The Sales
store = df.groupby('Store')['Sales'].sum()
store
```

```python
# Group By Multiple Columns
# Group By Category And Store And Sum the sales
multi = df.groupby(['Category','Store'])['Sales'].sum()
multi
```

### Aggregation
    Means to combine multiple things and have an answer just like mean

    Basic aggregation functions contains
    -Mean
    -Median
    -Mode
    -Standard Deviation
    -Varians
    

```python
df['Sales'].mean()
```

```python
df['Sales'].min()
df['Sales'].max()
df['Sales'].count()
df['Sales'].mode()
df['Sales'].median()
df['Sales'].std()
```

```python
df['Sales'].agg(['sum', 'mean', 'median', 'max', 'min'])
#not possible for modes
```
