### Merging 2 Dataframes

```python
# DataFrame 1: Employee information
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
# for merging dataframes there should be common in them like here we have employee_id
pd.merge(employees,salaries,on='employee_id',how="inner") #on means on what basis u want to merge
# the how thing by defaults use inner however u can see in the below cell the explaination of various joins
```

# Joins in merge (simple idea)

Think of 2 tables:
- Left table = employees
- Right table = salaries

You match them using employee_id.

1. Inner join
- Keeps only the rows that appear in both tables.
- Example: 1, 2, 3
- Memory: "only the overlap"

2. Left join
- Keeps all rows from the left table.
- Adds matching rows from the right table.
- If a left row has no match, it gets empty values.
- Example: 1, 2, 3, 4, 5
- Memory: "left side stays complete"

3. Right join
- Keeps all rows from the right table.
- Adds matching rows from the left table.
- Example: 1, 2, 3, 6, 7
- Memory: "right side stays complete"

4. Outer join
- Keeps all rows from both tables.
- Example: 1, 2, 3, 4, 5, 6, 7
- Memory: "everything from both"

5. Cross join
- Gives every row from the left table with every row from the right table.
- Memory: "all possible pairs"

Quick memory trick:
- Inner = overlap
- Left = left + overlap
- Right = right + overlap
- Outer = everything
- Cross = every combination

In pandas, this is controlled by:
- how='inner'
- how='left'
- how='right'
- how='outer'
- how='cross'

Note:
- The merge you already used is an Inner Join by default.

```python
pd.merge(employees,salaries,on='employee_id',how="outer") 
```

```python
pd.merge(employees,salaries,on='employee_id',how="left") 
# keeps the left column which is employee
```

```python
pd.merge(employees,salaries,on='employee_id',how="right") 
# keeps the right column based on salaries
```

### Concatination Of 2 Dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
# merging is done using concat function which merges two columns based on the column names
pd.concat([df2,df1])
```

```python
# now lets concat them in rows
pd.concat([df1,df2],axis=1)
```

### Joining Of 2 Dataframes

```python
df1 = pd.DataFrame({
    'name' : ['Alice','Bob','Charliw']
},index = [1,2,3])
df2 = pd.DataFrame({
    'Score' : [85,90,75]
},index= [2,3,4])
```

```python
df1.join(df2)
# it uses the same how factor inner outer left right the same thing
```

```python
df1.join(df2,how='outer')
```

```python
df1.join(df2,how='left')
```

```python
df1.join(df2,how='right')
```
