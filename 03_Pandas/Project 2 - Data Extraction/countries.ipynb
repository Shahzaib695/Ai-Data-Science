```python
# which country has the least population
df[df['population'] == df['population'].min()]['country']
```

```python
# what is the capital of the country with least population
df[df['population'] == df['population'].min()]['capital_city']
```

```python
# give me top 5 countries with top 5 highest democratic rate
df = df.sort_values(by = 'democracy_score',ascending=False)
```

```python
# how many regions are there
df['region'].value_counts().count()
```

```python
# how many countries lie in eastern europe region 
df['region'].value_counts()['Eastern Europe']
# another way
df[df['region'] == "Eastern Europe"]['country']
```

```python
# who is the political leader of the 2nd highest population country
# nlargest is used for finding the nth largest thing 
# who is the political leader of the 2nd highest population country
# nlargest is used for finding the nth largest thing 
# get the second-largest population value, then select the row(s) matching it
second_pop = df['population'].nlargest(2).iloc[1]
df[df['population'] == second_pop]['political_leader']
```

```python
# how many countries have unknown political leaders
uk = df[df['political_leader'].isna()]['country']
uk.count()
```

```python
# how many countries have republic in their name
count = 0
def hasrepublic(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt
df['country_long'].apply(hasrepublic)
print(count)
```
