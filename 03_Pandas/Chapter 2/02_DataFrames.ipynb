# Data Frames
    When we combine multiple series it becomes a data frame

```python
import pandas as pd
import numpy as np
```

### Creating A DataFrame

```python
data = {
    'Name': ['John','Anna','Peter','Linda'],
    'Age' : [20,34,19,20],
    'City' : ['New York','Paris','Berlin','London'],
    'Salary' : [45000,30000,25000,80000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#  we can use it also like using a list
```

```python
data_list2 = [
    ['John',28,'New York',60000],
    ['Anna',25,'Paris',80000],
    ['Peter',19,'Berlin',62000],
    ['Linda',20,'London',80000]
]
df2 = pd.DataFrame(data_list2)
df2
# the header disapear as u can see in the above sell we have the column names but no issue we can name them
columns = ['Name','Age','City','Salary']
df2 = pd.DataFrame(data_list2,columns =columns)
df2
```

### Removing Columns

```python
df2.drop('Designation',axis=1)
# axis = 1 means column by default the axis is zero 
```

```python
df2
#  here we can see the designation is present to fix this there is a keyword known as inplace
df2.drop('Designation',axis=1,inplace=True)
```

### Selecting Rows

```python
# To select rows we need to use loc and iloc [] to select multiple columns we need to add list inside of a list [[]]
df2.loc[0]
# here we are using 0 1 2 because the default index we have is 0 1 2 on the other hand if the index elements were different from 0 1 2 we could use them if i am not wrong i would request u to verify them. iloc on the other hand uses index values 0 1 2 3 even if the index values are not like this but it works like index of array 
```

```python
df2.loc[[0,1]]
```

```python
df2.iloc[3]
```

### Selecting Subsets Of Rows And Columns

    How to select particular portion of rows or columns to do this we have to select both rows and columns at the same tiem

```python
df2.loc[[0,1]][['City','Salary']]
# we have selected these columns
# df2.loc[[0,1]] -> this part selects the rows to select
# [['City','Salary']] -> this part selects the columns u want to see
# Practice set select row number 3 and 4. And for column name and age
```

```python
df2.loc[[2,3]][['Name','Age']]
```

```python
df2.iloc[[2,3]][['Name','City']]
```

```python
df2.loc[0:4, ['Name', 'City']]
# to select two or more rows we do this
```

### Conditional Selection

```python
# in pandas to target a condition we write like this
# data frame name first then brackets then again name then the column name then the condition just like this i have done an example below
df2[df2['Age'] >= 25]
```

```python
df2[df2['Age'] >= 25 & df2['City'] == "Paris"]
```

```python
df2[(df2['Age'] >= 25) & (df2['City'] == "Paris")] #to solve the above error we need to do is wrap the condition in round bracket
```
