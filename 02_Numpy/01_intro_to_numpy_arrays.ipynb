### Array Generation Functions 

```python
# we can generate array using arange function
arr = np.arange(1,11)
arr
arr = np.arange(2,15,2) #here 2 means with 2 gaps
arr
```

```python
# to generate an array with zeros it will generate vectors
arr = np.zeros(10)
arr
```

```python
# to generate an array with zeros. it will generate 2d dimensions 
# 12 columns in 2 rows
arr = np.zeros((12,2))
arr
```

```python
arr = np.ones(10) #this will generate vectors
arr
arr = np.ones((10,12))
arr
```

```python
# np.linspace generates evenly spaced numbers over a specified interval.
# Syntax: np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
# - start: first value
# - stop: last value (included by default)
# - num: number of samples to generate
# - endpoint: whether to include stop
# - retstep: if True, also return the spacing between values

# Example: generate 5 evenly spaced values from 0 to 12
x = np.linspace(0, 12, 3)
x
```

### Random Generation Function

```python
# np.random.rand generates samples from a uniform distribution over [0, 1)
# np.random.randn generates samples from a standard normal distribution (mean 0, variance 1)
#
# Differences:
# - rand: uniform distribution, values in [0, 1)
# - randn: normal (Gaussian) distribution, values can be negative or positive
# - rand uses a uniform distribution; randn uses a normal distribution
# - both accept shape arguments like np.random.rand(2, 3) or np.random.randn(2, 3)

u = np.random.rand(2, 3)
n = np.random.randn(2, 3)
u, n
```

```python
# numpy randint examples and brief notes:
# np.random.randint(low, high=None, size=None, dtype=int)
# - If high is None: returns integers in [0, low)
# - Otherwise returns integers in [low, high) (high is exclusive)
# - size can be an int or tuple for shape; dtype can be specified
# - For the newer Generator API prefer np.random.default_rng().integers(...)

# single integer in [0, 10)
r1 = np.random.randint(10)

# single integer in [5, 15)
r2 = np.random.randint(5, 15)

r3 = np.random.randint(10,40,12)
r3
```

```python
# Demonstrate np.sum with different axis values on the existing `arr`
print("arr:\n", arr)

# No axis (or axis=None): sum of all elements -> scalar
print("np.sum(arr):", np.sum(arr))

# axis=0: sum over rows -> column-wise sums (one value per column)
print("np.sum(arr, axis=0):", np.sum(arr, axis=0))

# axis=1: sum over columns -> row-wise sums (one value per row)
print("np.sum(arr, axis=1):", np.sum(arr, axis=1))
```

### Reshaping And Resizing

```python
arr = np.arange(1,31)
arr
```

```python
# .reshape changes the shape of an array without changing its data
# It is useful to convert a 1D vector into a 2D matrix.
print("original arr shape:", arr.shape)
print("original arr:", arr)

# Convert the 1D vector into a 2D matrix with 5 rows and 6 columns
matrix = arr.reshape(5, 6)
print("\nreshaped matrix (5, 6):")
print(matrix)
print("matrix shape:", matrix.shape)
# If the total number of elements does not match, reshape will raise an error
# For example, arr.reshape(4, 8) would fail because 4*8 != 30
```
