### Numpy Indexing And Slicing Of Vectors

```python
arr = np.arange(11,21)
arr
```

```python
# same as list and array like in other langauges
arr[6]
```

```python
arr[9]
```

```python
# to print a sequence of number like from 11-16 we will be writing like this [starting index: ending index (n-1)]. By default the starting value is 0 so if we dont write anything on starting point it will assume 0 by default. And for the ending point if we leave that so by default the value will be the last index
arr[0:6]
arr[3::2] #will print from 14-20 ::-> means steps 
```

### Numpy Indexing And Slicing Of Matrix

```python
#  lets create a 2d array using arrange function
arr_matrix = np.arange(1,31).reshape(6,5)
arr_matrix
```

```python
# if we will do the same way like we did for vectors it will print the complete row but to print an appropriate index like we have to print the number 2 we can write like this [row index it also starts from 0 the first one will be 0 and the second row will be 1 and it will be in the same,the index of the element in that appropriate row]
arr_matrix[0] # this will print the first row
arr_matrix[0,4] #this will print the 2nd element at index 1 in the first row 
```

```python
# to print a elements from both column 1 and 2 we will be writing like this as below [column start point:endpoint of the second column remember n-1, then the element number u want to print from both column and then the end point of the second element remember n-1]
arr_matrix[0:2,1:3]
```

```python
arr_matrix[3:6,3:6]
```

```python
# now print the column number 3
arr_matrix[:,2]
# select the rows as u will be requiring all rows leaving that empty by default takes the start point and end point and after comma u will write the row u want to print from each column
```
