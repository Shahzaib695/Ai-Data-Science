#### Broadcasting


```python
# Broadcasting is when NumPy automatically expands a smaller array to match a larger one
# so you can do elementwise operations without having the exact same shape.
# For example, adding 1 to a1 adds 1 to every element in a1.
a1 = np.arange(1,11)
print("Broadcasting means NumPy 'stretches' smaller arrays to match larger ones for operations.")
print("a1 + 1 ->", a1 + 1)
```

```python
# this also works on matrix
a2 = np.arange(1,26).reshape(5,5)
a2
```

```python
a2 + 10 #adds 10 to every number
```

#### Deep And Shallow Copy

    Shallow Copy
    It means that if u have sliced out a part from array and ur changing it it will not affect the original array 

```python
a = np.arange(1,12)
a
```

```python
slice = a[:5]
slice
```

```python
# now lets multiply the sliced part using broadcast approach
slice = slice * 9
slice
```

```python
# here u can see the original array stayed intact
print(f"Original Array -> {a}")
print(f"Sliced Array -> {slice}")
```

    Deep Copy

#### Advance Array Manipulation


    Stacking An Array
    It means to stack arrays with each other. You can not stack matrix but u can stack vectors.
    There are 3 types of stack Vertical,Horizontal and Column Stack

```python
a = np.arange(1,6)
b = np.arange(6,11)
```

```python
b
```

```python
np.vstack((a,b))
# in line 
```

```python
np.hstack((a,b)) 
#row wise stack hogaya
```

```python
np.column_stack((a,b)) 
# stacks each array as different columns
```
