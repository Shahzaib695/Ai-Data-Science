### Subplots in Matplotlib

### Working 
    In this we have to give rows and columns. For example if we wrote plt.subplot(2,2) it means we will have 2 rows and 2 columns while plotting we work on something known as canvas if we want to target the first column of the first row we can write like this plt.subplot(2,2,1).
    No restriction of how many rows or columns u want u can even write 6,6 means u will have 6 rows and 6 columns

```python
# if u want to plot just add the plot beneath the subplot u want to plot
plt.subplot(2,2,1)
plt.plot(x**2,y**3)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
```

### Object Oriented Way

    As we saw previously hwo we could create subplots on a canvas but the core issue was that we werent able to select that canvas for which we have an object oriented way in this way we can have a control over our canvas

```python
fig = plt.figure()
```

    "fig.add_axes([left, bottom, width, height])\n"
    "- left, bottom: distance from the figure's left and bottom edges (fractions 0.0 - 1.0)\n"
    "- width, height: size of the axes as fractions of the figure (0.0 - 1.0)\n"
    "All four numbers are in figure coordinates (fractional), e.g. [0.1, 0.1, 0.8, 0.8]."

```python
# 0.1 means 10 percent 0.1-1 represents 10-100% height
# first two means for left side and for the next two sides its for right side
fig = plt.figure(figsize=(12,8),dpi=200)
axis1 = fig.add_axes([0.1,0.1,1,1]) # type: ignore
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.6,.7]) #type: ignore
axis2.plot(x,y)
#left , bottom , right,top
```
