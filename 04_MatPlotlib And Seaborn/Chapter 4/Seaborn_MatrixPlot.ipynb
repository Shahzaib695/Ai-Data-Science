```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Heat Map

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

    To create a heat map first we need to know what is co relation. We use co relation to create heat map.
* Co-Relation:
    Suppose there are 3 columns named A,B,C and there is a realtionship between A and C so we can use a to predict c. The co relation comes from -1 to 1 and in the middle element we have 0. If the value is zero of co relation then it means the variable is not suitable for prediction. There can be a positive co relation as well as negative co relation.
* Positive Co-Relation:
    If the value of column a is increasing and the value of c is increasing it is a sign of positive co-relation.
* Negative Co-Relation:
    Now look at column b and c if the value of b is decreasing the value of c is increasing this is a sign of negative co-relation.

### Reference Column:
    For reference of the column this is the exact table that we are using for explaination


This is an important concept for machine learning if we want to show this relation on a plot we use heat map

```python
tips_corr = tips[['total_bill','tip','size']]
tips_corr
```

```python
# lets calculate the co relation between these
tips_corr.corr()
```

```python
sns.heatmap(tips_corr.corr())
```

#### What is happening??
    The brighter the color the more it shows the co relation between elements 

```python
sns.heatmap(tips_corr.corr(),annot=True)
```

### Cluster Map

```python
sns.clustermap(tips_corr.corr())
```
