```python
import numpy as np
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
# to remove the outliers we need first to find out IQR to find iqr we need to find out what our Q1,Q3 are in the array
Q1 = np.percentile(arr,25)
Q2 = np.percentile(arr,50)
Q3 = np.percentile(arr,75)
# 25 percentile -> Q1
Q1,Q2,Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
# Now find UF and LF so that we can see what values are outliers easily
UF = Q3 + (1.5*(IQR))
LF = Q1 - (1.5*(IQR))
UF,LF
# The numbers which are not lying in the range between lf and uf we will remove them
```

```python
l = []
for i in arr:
    if i >= LF and i <= UF:
        l.append(i)
l
# we successfully removed our outlier
arr2 = np.array(l) # -> outlier removed
arr,arr2
```

```python
import seaborn as sns
sns.boxplot(x=arr)
# the dot on the right side it represent outlier this is the boxplot with an outlier and due to that sole reason the box plot is so odd
```
